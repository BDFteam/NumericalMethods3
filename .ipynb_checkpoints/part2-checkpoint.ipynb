# Part 2: Bivariate optimization
First import dependencies.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from random import randint

import warnings
```

```python
rosenburg = lambda x0, x1: 100 * ((x1-x0**2)**2) + (1 - x0)**2
rosenburg_grad = lambda x0, x1: np.array([-400*x0*(x1-x0**2)-2*(1-x0), 200*(x1 - x0**2)])
rosenburg_hessian = lambda x0, x1: np.array([[-400*(x1-x0**2) + 800 * x0**2 + 2, -400*x0],[-400*x0, 200]])


rastrigin = lambda x0, x1: (x0 ** 2 + 10 * (np.cos(2 * np.pi * x0))) + (x1 ** 2+ 10 * (1 - np.cos(2 * np.pi * x1)))
rastrigin_grad = lambda x0, x1: np.array([2*x0 + 20*np.pi*np.sin(2*np.pi*x0), 2*x1 + 20*np.pi*np.sin(2*np.pi*x1)])
rastrigin_hessian = lambda x0, x1: np.array([[2 + 40*(np.pi**2)*np.cos(2*np.pi*x0), 0],[0, 2 + 40*(np.pi**2)*np.cos(2*np.pi*x1)]])
```

## Functions
Define all the functions we'll need. We have that the Rosenburg function is:
$$f_1(x_0, x_1) = 100(x_1 - x_0^2)^2 + (1 - x_0)^2$$
Which has the following Gradient and Hessian:
$$\nabla f_1(x_0, x_1) = \begin{bmatrix} -400 x_0 (x_1 - x_0^2) - 2(1-x_0) \\ 200(x_1 - x_0^2) \end{bmatrix}$$
$$\nabla^2 f_1(x_0, x_1) = \begin{bmatrix} -400 (x_1 - x_0^2) + 800 x_0^2  + 2 && -400 x_0 \\ -400 x_0 && 200 \end{bmatrix}$$

And we also have the Rastrigin function:
$$f_2(x_0, x_1) = (x_0^2 + 10(1- \cos(2\pi x_0))) + (x_1^2 + 10(1 - \cos(2\pi x_1)))$$

Which has the following Gradient and Hessian:
$$\nabla f_2(x_0, x_1) = \begin{bmatrix} x_0 + 20\pi \sin(2\pi x_0) \\ x_1 + 20\pi \sin(2\pi x_1) \end{bmatrix}$$
$$\nabla^2 f_2(x_0, x_1) = \begin{bmatrix} 2 + 40\pi^2\cos(2\pi x_0) && 0 \\ 0 && 2 + 40\pi^2\cos(2\pi x_1) \end{bmatrix}$$

## Newton-Rapson method
Below we create the Newton-Rapson method, which has the following stopping conditions: if the change per iteration is below a certain $\epsilon_1$ (default: $\epsilon_1 = 0.0001$), if it reaches it maximum number of iterations (default is 10.000), or if the norm of the step size is below an $\epsilon_2$ (default: $\epsilon_2 = 0.001$).

```python
def mini_NR(func, x0, gradient, hessian, iteration_err = 0.0001, max_iterations = 10000, min_change = 0.001, advanced = False):
    x: list = [np.inf, x0]
    i = 0
    
    while np.abs(np.linalg.norm(x[-1] - x[-2])) > iteration_err and i < max_iterations or (advanced is True and i < max_iterations):
        x.append(x[-1] - np.linalg.inv(hessian(x[-1][0], x[-1][1])).dot(gradient(x[-1][0], x[-1][1])))
        i += 1

        if np.linalg.norm(np.linalg.inv(hessian(x[-1][0], x[-1][1])).dot(gradient(x[-1][0], x[-1][1]))) < min_change and advanced is False:
            break

    print(f"Newton Iterations: {i}")

    if advanced:
        return np.array(x)

    return x[-1]
```

## Differential Evolution
Below we create the differential evolution method. There are two stopping conditions: maximum number of generations, or little diversity in the population (i.e. variance is below value).

```python
def randint_multiple(start, stop, size):
    """Generate random int"""
    random_ints = []
    for _ in range(size):
        random_ints.append(randint(start, stop))

    return random_ints

def generate_rand_bound(bound, size):
    """Generate random values in bound"""
    length = np.abs(bound[0] - bound[1])
    return np.add(np.random.rand(size) * length, bound[0])

def generate_random_points(bound, size):
    """The most inefficient way to create random points in a bound!"""
    a = []
    for b in bound:
        a.append(generate_rand_bound(b, size))
        
    c = np.zeros((size, bound.shape[0]))

    for i in range(size):
        for j in range(bound.shape[0]):
            c[i][j] = a[j][i]

    return c

def get_values(func, estimations):
    """Calculate values of estimations"""
    values = np.zeros(estimations.shape[0])

    for i in range(estimations.shape[0]):
        values[i] = func(estimations[i][0], estimations[i][1])

    return values

def mini_de(func, bound, popsize, mutation, max_generations = 100, max_std = 0.00000001, advanced = False, return_iterations=False):
    """Minimize"""
    iterations = 0
    # init with random points from bound.
    x = [generate_random_points(bound, popsize)]
    for i in range(1, max_generations):
        z = np.zeros((popsize, bound.shape[0]))

        for j in range(popsize):
            r0, r1, r2 = randint_multiple(0, popsize - 1, 3)
            z[j] = np.add(x[-1][r0], (x[-1][r1] - x[-1][r2]) * mutation)

            if not (func(z[j][0], z[j][1]) <= func(x[-1][j][0], x[-1][j][1])):
                z[j] = x[-1][j]

        x.append(z)

        if np.std(get_values(func, x[-1])) < max_std:
            if return_iterations is False:
                print(f"DE iterations: {i}")
            iterations = i
            break

        if i == max_generations -1:
            iterations = i
            if return_iterations is False:
                print(f"DE evolutions: {i}")

    if advanced:
        z = []
        for a in x:
            z.append(a[np.argmin(get_values(func, a))])
        return np.array(z)

    if return_iterations:
        return x[-1][np.argmin(get_values(func, x[-1]))], iterations

    return x[-1][np.argmin(get_values(func, x[-1]))]
```

### Population size
We can test how the population size impacts the convergence speed, and quality.

```python
def population_change():
    n = 50

    true_value = np.array([1, 1])
    print(f"True: {true_value}")
    print("")
    
    print("Population 10:")
    pop10 = np.zeros((n, 2))
    iter10 = np.zeros(n)
    for i in range(n):
        pop10[i], iter10[i] = mini_de(rosenburg, np.array([[-5, 5], [-5, 5]]), 10, 0.8, max_generations=1000, return_iterations=True)
    print(f"x Mean: {pop10[:][0].mean()}, x Est. std.: {pop10[:][0].std()}, Avg. Iterations: {iter10.mean()}")
    print("")

    print("Population 25:")
    pop25 = np.zeros((n, 2))
    iter25 = np.zeros(n)
    for i in range(n):
        pop25[i], iter25[i] = mini_de(rosenburg, np.array([[-5, 5], [-5, 5]]), 25, 0.8, max_generations=1000, return_iterations=True)
    print(f"x Mean: {pop25[:][0].mean()}, x Est. std.: {pop25[:][0].std()}, Avg. Iterations: {iter25.mean()}")
    print("")

    print("Population 50:")
    pop50 = np.zeros((n, 2))
    iter50 = np.zeros(n)
    for i in range(n):
        pop50[i], iter50[i] = mini_de(rosenburg, np.array([[-5, 5], [-5, 5]]), 50, 0.8, max_generations=1000, return_iterations=True)
    print(f"x Mean: {pop50[:][0].mean()}, x Est. std.: {pop50[:][0].std()}, Avg. Iterations: {iter50.mean()}")
    print("")

    print("Population 100:")
    pop100 = np.zeros((n, 2))
    iter100 = np.zeros(n)
    for i in range(n):
        pop100[i], iter100[i] = mini_de(rosenburg, np.array([[-5, 5], [-5, 5]]), 100, 0.8, max_generations=1000, return_iterations=True)
    print(f"x Mean: {pop100[:][0].mean()}, x Est. std.: {pop100[:][0].std()}, Avg. Iterations: {iter100.mean()}")
    print("")

    print("Population 250:")
    pop250 = np.zeros((n, 2))
    iter250 = np.zeros(n)
    for i in range(n):
        pop250[i], iter250[i] = mini_de(rosenburg, np.array([[-5, 5], [-5, 5]]), 250, 0.8, max_generations=1000, return_iterations=True)
    print(f"x Mean: {pop250[:][0].mean()}, x Est. std.: {pop250[:][0].std()}, Avg. Iterations: {iter250.mean()}")
    print("")

    print("Population 1000:")
    pop1000 = np.zeros((n, 2))
    iter1000 = np.zeros(n)
    for i in range(n):
        pop1000[i], iter1000[i] = mini_de(rosenburg, np.array([[-5, 5], [-5, 5]]), 250, 0.8, max_generations=1000, return_iterations=True)
    print(f"x Mean: {pop1000[:][0].mean()}, x Est. std.: {pop1000[:][0].std()}, Avg. Iterations: {iter1000.mean()}")
    print("")

population_change()
```

We can clearly see that for small poplulations, a lot of iterations are needed. However after populations of 25 we can see that the average iterations needed increase again. This makes sense, as there are more points and the main stopping condition is that the standard deviation is below a certain threshold. However if we look at the std. deviation of the estimation, it seems that the estimations with larger population sizes are much closer together (and also closer to the true value).

# Rosenberg

```python
def rosenburg_min():
    x0 = np.array([5, 5])
    x_min_newton = mini_NR(rosenburg, x0, rosenburg_grad, rosenburg_hessian, 0.00001)
    x_min_be = mini_de(rosenburg, np.array([[-5, 5], [-5, 5]]), 100, 0.8)

    print(f"Newton min: {x_min_newton}")
    print(f"Be min: {x_min_be}")

rosenburg_min()
```

## Rastrigin

```python
def rastrigin_min():
    x0=np.array([0.01, 0.01])
    x_min_newton = mini_NR(rastrigin, x0, rastrigin_grad, rastrigin_hessian, 0.000001)
    x_min_de = mini_de(rastrigin, np.array([[-5, 5], [-5, 5]]), 100, 0.8)

    print(f"Newton min: {x_min_newton}")
    print(f"DE min: {x_min_de}")

rastrigin_min()
```

## Convergence plot

```python
def distance(est, true):
    retur = np.zeros(est.shape[0])
    for i, e in enumerate(est):
        retur[i] = np.linalg.norm(e - true)

    return retur
 
def convergence_plot():
    x0 = np.array([5, 5])
    warnings.filterwarnings('ignore')
    y_min_newton = mini_NR(rosenburg, x0, rosenburg_grad, rosenburg_hessian, 0, max_iterations=25, advanced=True)
    y_min_de = mini_de(rosenburg, np.array([[-5, 5], [-5, 5]]), 100, 0.8, advanced=True)

    true = np.array([1, 1])

    x = np.linspace(0, 20, 20)
    plt.plot(x, distance(y_min_newton, true)[1:21])
    plt.plot(x, distance(y_min_de, true)[:20])
    plt.yscale('log')
    plt.xticks([0, 5, 10, 15, 20])
    plt.show()

convergence_plot()
```

Clearly you can see that the Newtons Method converges much quicker then then the Differential Evolution method.
